```python
import pandas as pd
BOS_df1 = pd.read_excel('Bos_excel.xlsx', sheet_name=20, header=5)
BOS_df1 = BOS_df1.dropna()
BOS_df1['Narrative'] = BOS_df1['Description\nNarrative'].str.extract('(SECURITY .*)$',expand=True)
BOS_df1['Description'] = BOS_df1['Description\nNarrative'].str.extract('(.*)',expand=True)
BOS_df1['Description1'] = BOS_df1['Description\nNarrative'].str.extract('^.*(\n.*)',expand=True)
BOS_df1 = BOS_df1.replace({
    'Description1':'\n'
},'',regex=True)
BOS_df1 = BOS_df1.replace({
    'Description1':['SECURITY PURCHASE', 'SECURITY SALE']
},'')
BOS_df1['Description'] = BOS_df1['Description']+BOS_df1['Description1']
BOS_df1.drop(columns = ['Description\nNarrative', 'Description1'], inplace = True)
BOS_df1['Value Date'] = BOS_df1['Trade Date\nValue Date'].str.extract('.*\n(.*)$',expand=True)
BOS_df1['Trade Date'] = BOS_df1['Trade Date\nValue Date'].str.extract('(.*)',expand=True)
BOS_df1.drop(columns = ['Trade Date\nValue Date'], inplace = True)
BOS_df1 = BOS_df1.reindex(columns = ['Ccy','Description','Narrative','Quantity/Nominal','Reference Number','Trade Date','Value Date'])
BOS_df1['Trade Date'] = pd.to_datetime(BOS_df1['Trade Date']).dt.strftime('%m/%d/%Y')
BOS_df1['Value Date'] = pd.to_datetime(BOS_df1['Value Date']).dt.strftime('%d/%m/%Y')
BOS_Temp_df=BOS_df1.iloc[[2,4,5,10],[3]]
BOS_Temp_df['Quantity/Nominal']=BOS_Temp_df['Quantity/Nominal'].map(lambda x: x.lstrip('(').rstrip(')').replace(',' , ''))
BOS_df1 = pd.merge(BOS_Temp_df,BOS_df1, left_index=True, right_index=True, on='Quantity/Nominal', how='outer')
BOS_df1['Quantity/Nominal'] = pd.to_numeric(BOS_df1['Quantity/Nominal'], downcast = 'float')
BOS_df1 = BOS_df1.reindex(columns = ['Ccy','Description','Narrative','Quantity/Nominal','Reference Number','Trade Date','Value Date'])
BOS_df1
```

```python
BOS_df2 = pd.read_excel('Bos_excel.xlsx', sheet_name=21,header=4,skip_footer=14)
BOS_df2 = BOS_df2.dropna(axis=1,how='all')
BOS_df2.fillna(value='', inplace=True)
BOS_df2.ix[0] = BOS_df2.ix[0]+BOS_df2.ix[1]
header = BOS_df2.iloc[0]
BOS_df2 = BOS_df2[1:]
BOS_df2 = BOS_df2.rename(columns = header)
BOS_df2['Quantity/Nominal']= BOS_df2['Quantity/Nominal'].map(lambda x: str(x))
BOS_df2.ix[2] = BOS_df2.ix[2]+'\n'+BOS_df2.ix[3]+'\n'+BOS_df2.ix[4]
BOS_df2.ix[5] = BOS_df2.ix[5]+'\n'+BOS_df2.ix[6]+'\n'+BOS_df2.ix[7]
BOS_df2.reset_index(drop=True,inplace=True)
BOS_df2.drop(BOS_df2.index[[0,2,3,5,6]],inplace=True)
BOS_df2.reset_index(drop=True,inplace=True)
BOS_df2.replace({
    'Ccy': '\n\n',
    'Quantity/Nominal': '\n\n',
    'Reference Number': '\n\n'
}, '', regex=True, inplace=True)
BOS_df2['Value Date'] = BOS_df2['Trade DateValue Date'].str.extract('.*\n+(.*)$',expand=True)
BOS_df2['Trade Date'] = BOS_df2['Trade DateValue Date'].str.extract('(.*)',expand=True)
BOS_df2['Narrative'] = BOS_df2['DescriptionNarrative'].str.extract('(SECURITY .*)$',expand=True)
BOS_df2['Description'] = BOS_df2['DescriptionNarrative'].str.extract('(.*)',expand=True)
BOS_df2['Description1'] = BOS_df2['DescriptionNarrative'].str.extract('^.*(\n.*)',expand=True)
BOS_df2 = BOS_df2.replace({
    'Description1':'\n'
},'',regex=True)
BOS_df2 = BOS_df2.replace({
    'Description1':['SECURITY PURCHASE', 'SECURITY SALE']
},'')
BOS_df2['Description'] = BOS_df2['Description']+BOS_df2['Description1']
BOS_df2.drop(columns = ['DescriptionNarrative', 'Description1', 'Trade DateValue Date'], inplace = True)
BOS_df3 = pd.concat([BOS_df1,BOS_df2], ignore_index=True)
BOS_df3
```
