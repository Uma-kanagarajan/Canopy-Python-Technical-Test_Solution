```python
import pandas as pd
JBS_df = pd.read_excel("JBS_PG7.xlsx",header=4, skip_footer=5)
JBS_df.fillna(value='', inplace=True)
JBS_df['Debit']= JBS_df['Debit'].map(lambda x: str(x))
JBS_df['Credit']= JBS_df['Credit'].map(lambda x: str(x))
JBS_df['Balance']= JBS_df['Balance'].map(lambda x: str(x))
JBS_df.iloc[3] = JBS_df.iloc[3]+JBS_df.iloc[4]+JBS_df.iloc[5]+JBS_df.iloc[6]
JBS_df.drop(JBS_df.index[[0,1,4,5,6]],inplace=True)
JBS_df.reset_index(drop=True,inplace=True)
JBS_df['Credit']=pd.to_numeric(JBS_df['Credit'], downcast='float')
JBS_df['Debit']=pd.to_numeric(JBS_df['Debit'],downcast='float')
JBS_df['Balance']=pd.to_numeric(JBS_df['Balance'],downcast='float')
JBS_df['Txn Date1'] = JBS_df['Txn Date'].str.extract('.*n(.*)$',expand=True)
JBS_df['Txn Date1'] = JBS_df['Txn Date'].str.extract('[0-9.]*(.*)$',expand=True)
JBS_df['Txn Date'] = JBS_df['Txn Date'].str.extract('([0-9.]*)',expand=True)
JBS_df['Txn Date1'] = JBS_df['Txn Date1'].replace('n','',regex=True)
JBS_df['Booking Text']= JBS_df['Booking Text']+JBS_df['Txn Date1']
JBS_df.drop(columns = ['Txn Date1'],inplace=True)
JBS_df['Booking Date'] = pd.to_datetime(JBS_df['Booking Date']).dt.strftime('%d/%m/%Y')
JBS_df['Txn Date'] = pd.to_datetime(JBS_df['Txn Date']).dt.strftime('%d/%m/%Y')
JBS_df['Value Date'] = pd.to_datetime(JBS_df['Value Date']).dt.strftime('%d/%m/%Y')
JBS_df.fillna(value={
    'Debit':0,
    'Credit':0
})
```
